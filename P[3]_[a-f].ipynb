ساخت دیکشنری لغات و نمودار برای 10 کلمه پرتکرار در کورپس

````python
import json
````

````python
def dictionary(token):
        word_count = {}
        # print(self.token)
        for j in token:
            for word in j:
                if word not in word_count:
                    word_count[word] = 1
                elif word in word_count:
                    word_count[word] += 1

        sorteddict = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
        converted_dict = dict(sorteddict)
        with open('convert.txt', 'w', encoding='utf-8') as convert_file:
            convert_file.write(json.dumps(converted_dict, ensure_ascii=False))
        return converted_dict

    # ----->   bar chart 10 first frequent words
def barChart(token):
        converted_dict = dictionary(token)
        names = list(converted_dict.keys())[0:10]
        values = list(converted_dict.values())[0:10]
        plt.bar(range(10), values, tick_label=names)
        plt.show()
````

````python
res=dictionary(token)

word=list(res.keys())[:5]
dup=list(res.values())[:5]
print(word,dup)
````
